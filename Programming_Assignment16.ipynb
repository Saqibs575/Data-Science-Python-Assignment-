Question1. Write a function that stutters a word as if someone is struggling to read it. The
first two letters are repeated twice with an ellipsis ... and space after each, and then the
word is pronounced with a question mark ?.<BR><BR>
**Examples**<BR><BR>
```
stutter("incredible") ➞ "in... in... incredible?"
stutter("enthusiastic") ➞ "en... en... enthusiastic?"
stutter("outstanding") ➞ "ou... ou... outstanding?"
```

````python
def stutter(word) :

    return 2*(word[:2]+'...')+word+'?'

print(stutter('incredible'))
print(stutter('enthusiastic'))
print(stutter('outstanding'))
````

Question 2.Create a function that takes an angle in radians and returns the corresponding
angle in degrees rounded to one decimal place.<br><BR>
**Examples**<BR><BR>
```
radians_to_degrees(1) ➞ 57.3
radians_to_degrees(20) ➞ 1145.9
radians_to_degrees(50) ➞ 2864.8
```

````python
import math
def radians_to_degree(r) :
    return round(r*180 / math.pi,1)

print(radians_to_degree(1))
print(radians_to_degree(20))
print(radians_to_degree(50))
````

Question 3. In this challenge, establish if a given integer num is a Curzon number. If 1 plus
2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon
number.<BR><BR>
Given a non-negative integer num, implement a function that returns True if num is a Curzon
number, or False otherwise.<BR><BR>
**Examples**<BR><BR>
```
is_curzon(5) ➞ True
# 2 ** 5 + 1 = 33
# 2 * 5 + 1 = 11
# 33 is a multiple of 11

is_curzon(10) ➞ False
# 2 ** 10 + 1 = 1025
# 2 * 10 + 1 = 21
# 1025 is not a multiple of 21

is_curzon(14) ➞ True
# 2 ** 14 + 1 = 16385
# 2 * 14 + 1 = 29
# 16385 is a multiple of 29
```

````python
def is_curzon(num) :

    if (num >= 0 ) and ((2**num+1) % (2*num +1) == 0) :
        return True
    return False

print(is_curzon(5))
print(is_curzon(10))
print(is_curzon(14))
````

Question 4.Given the side length S find the area of a hexagon.<BR><BR>Screenshot 2022-10-15 112533.png<br><BR>
**Examples**<br><BR>
```
area_of_hexagon(1) ➞ 2.6
area_of_hexagon(2) ➞ 10.4
area_of_hexagon(3) ➞ 23.4
```

````python
import math 
def area_of_hexagon(S) :

    return round(3*math.sqrt(3)*pow(S,2)/2,1)

print(area_of_hexagon(1))
print(area_of_hexagon(2))
print(area_of_hexagon(3))
````

Question 5. Create a function that returns a base-2 (binary) representation of a base-10
(decimal) string number. To convert is simple: ((2) means base-2 and (10) means base-10)
010101001(2) = 1 + 8 + 32 + 128.<BR><BR>
Going from right to left, the value of the most right bit is 1, now from that every bit to the left
will be x2 the value, value of an 8 bit binary numbers are (256, 128, 64, 32, 16, 8, 4, 2, 1).<BR><BR>
**Examples**<BR><BR>


```
binary(1) ➞ "1"
# 1*1 = 1

binary(5) ➞ "101"
# 1*1 + 1*4 = 5

binary(10) ➞ "1010"
# 1*2 + 1*8 = 10

```



````python
def binary(number) :
    bin = ""
    decimal = int(number)
    
    while decimal > 0 :
        bin = str(decimal % 2) +bin
        decimal = decimal //2
    
    return bin

print(f'"{binary(1)}"')
print(f'"{binary(5)}"')
print(f'"{binary(10)}"')
````
