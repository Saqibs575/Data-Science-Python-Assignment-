Q1.<BR> Write a program that calculates and prints the value according to the given formula:<BR>
Q = Square root of [(2 * C * D)/H]<br>
Following are the fixed values of C and H:<br>
C is 50. H is 30.<br>
D is the variable whose values should be input to your program in a comma-separated
sequence.<BR>
Example<br>
Let us assume the following comma separated input sequence is given to the program:<BR>
100,150,180<BR>
The output of the program should be:
<BR>18,22,24

```python
def calc(D) :
    C = 50 
    H = 30
    Q = (2*C*D/H)**0.5
    return int(Q)

print('Q = sqrt[(2CD)/H]\n')
D = map(int,input('Give comma separated value of D : ').split(','))
print('\nQ = ',end = '')

output = [f"{calc(i)}" for i in D ]
print(','.join(output))
```

Q3.<br>Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The
element value in the i-th row and j-th column of the array should be i*j.<br>
Note: i=0,1.., X-1; j=0,1,Â¡Y-1.<br>
Example<br>
Suppose the following inputs are given to the program:<BR>
3,5<br>
Then, the output of the program should be:
<br>[[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]]

```python
def array(X,Y) : 
    array = []

    for i in range (X) :
        a = []
        for j in range (Y) :
            a.append(i*j)
        array.append(a)

    return array

X = int(input('X = ')) 
Y = int(input('Y = '))
print(array(X,Y))
```

Q4<BR>
Write a program that accepts a comma separated sequence of words as input and prints the<BR>
words in a comma-separated sequence after sorting them alphabetically.<br>
Suppose the following input is supplied to the program:<br>
without,hello,bag,world<br>
Then, the output should be:<br>
bag,hello,without,world<br>

```python
def alphabetic(sequence) :
    sequence=sequence.lower().split(',')
    return ','.join(sorted(sequence))

sequence = input('Give Comma separated sequence : ')
print('Required output is : ',alphabetic(sequence))
```

Q5<br>
Write a program that accepts a sentence and calculate the number of letters and digits.<BR><BR>
Suppose the following input is supplied to the program:<BR><BR>
hello world! 123<BR><BR>
Then, the output should be:
<br><br>LETTERS 10<BR><BR>WORDS  10

```python
def countLetters(sequence) :
    sequence1 = list(sequence)
    count = 0

    for ch in sequence :
        if ch.isalpha() :
            count += 1
    return count 

def countDigits(sequence) :
    sequence1 = list(sequence)
    count = 0

    for ch in sequence :
        if ch.isdigit() :
            count += 1
    return count

sentence = input ('Give your sentence : ')
print('\nLEETERS  ',countLetters(sentence))
print(" DIGITS  ",countDigits(sentence))
```

Q6<br>
A website requires the users to input username and password to register. Write a program to
check the validity of password input by users.<br><br>
Following are the criteria for checking the password:
1. At least 1 letter between [a-z]
2. At least 1 number between [0-9]
1. At least 1 letter between [A-Z]
3. At least 1 character from [$#@]
4. Minimum length of transaction password: 6
5. Maximum length of transaction password: 12<br><BR>
Your program should accept a sequence of comma separated passwords and will check them
according to the above criteria. Passwords that match the criteria are to be printed, each
separated by a comma.<BR><BR>
Example<BR><BR>
If the following passwords are given as input to the program:<BR><BR>
ABd1234@1,a F1#,2w3E*,2We3345<BR><BR>
Then, the output of the program should be:<BR><BR>
ABd1234@1

```python
def isValid(password) :
    flag1 ,flag2,flag3,flag4 = False,False,False,False
    splChar = '$#@'

    if (6 <= len(password) <= 12) :

        for ch in password :

            if ch.islower() :
                flag1 = True 
                
            elif ch.isupper() :
                flag2 = True 
                
            elif ch.isdigit() :
                flag3 = True 
                
            elif (ch in splChar) :
                flag4 = True

        if flag1 == True and flag2 == True and flag3 == True and flag4 == True :
            return True
    return False

Input = input('Give Comma separated passwords : ').split(',')
for password in Input :
    if isValid(password) :
        print(f'\n{password}')
```
